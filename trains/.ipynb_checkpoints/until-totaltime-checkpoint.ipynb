```python
# dataframe for Rajdhani
df2 = df.loc[df.type == "Raj"]
df2
```

```python
df2_1 = df2.loc[df2.from_station_code == "NDLS"]
df2_1
```

```python
plt.figure(figsize=(20,13))
plt.title("Rajdhani trains emerging from New Delhi", fontdict={'fontsize': 25})
plt.xlabel("Destination", fontdict={'fontsize': 20})
plt.ylabel("Distance covered", fontdict={'fontsize': 20})

plt.bar(df2_1.to_station_name, df2_1.distance)
plt.show()
```

```python
plt.figure(figsize=(20,13))
plt.title("Time taken by Rajdhani from New Delhi to Destination", fontdict={'fontsize': 25})
plt.xlabel("Destination", fontdict={'fontsize': 20})
plt.ylabel("Time taken", fontdict={'fontsize': 20})

plt.bar(df2_1.to_station_name, df2_1["total time"], color="violet")
plt.show()
```

```python
plt.figure(figsize=(20,13))
plt.title("Average Speed of Rajdhani from New Delhi to Destination", fontdict={'fontsize': 25})
plt.xlabel("Destination", fontdict={'fontsize': 20})
plt.ylabel("Average Speed", fontdict={'fontsize': 20})

plt.bar(df2_1.to_station_name, df2_1["avg_speed"], color="green")
plt.show()
```

```python
indices = df[df.distance == 0].index
```

```python
df = df.drop(indices)
```

```python
df["avg_speed"] = df.distance/df["total time"]
```

```python
df.head(50)
```

```python
x = df.avg_speed.mean()
y = df.avg_speed.median()

if x > y:
    print("-> Right skewed")
else:
    print("-> Left skewed")
    
print(f"Mean: {x}")
print(f"Median: {y}")
```

```python
plt.figure(figsize=(20,13))

bins = [x for x in range(10, 121, 10)]

plt.title("Distribution of Average Speed of Trains", fontdict={'fontsize':25})
plt.xlabel("Average Speed (kmph)", fontdict={'fontsize':20})
plt.ylabel("Number of Trains", fontdict={'fontsize':20})

plt.hist(df.avg_speed, bins=bins)

plt.show()
```

```python
x = df.distance.mean()
y = df.distance.median()

if x > y:
    print("-> Right skewed")
else:
    print("-> Left skewed")
    
print(f"Mean: {x}")
print(f"Median: {y}")
```

```python
plt.figure(figsize=(20,13))

bins = [x for x in range(0, 2001, 100)]

plt.title("Distribution of Distance travelled by Trains", fontdict={'fontsize':25})
plt.xlabel("Distance (km)", fontdict={'fontsize':20})
plt.ylabel("Number of Trains", fontdict={'fontsize':20})

plt.hist(df.distance, bins=bins)

plt.show()
```

```python
x = df["total time"].mean()
y = df["total time"].median()

if x > y:
    print("-> Right skewed")
else:
    print("-> Left skewed")
    
print(f"Mean: {x}")
print(f"Median: {y}")
```

```python
plt.figure(figsize=(20,13))

bins = [x for x in range(0, 90)]

plt.title("Distribution of Time taken by Trains", fontdict={'fontsize':25})
plt.xlabel("Time taken (hrs)", fontdict={'fontsize':20})
plt.ylabel("Number of Trains", fontdict={'fontsize':20})

plt.hist(df["total time"], bins=bins)

plt.show()
```

```python
plt.title("Distribution of running time of Different trains in India", fontdict={'fontsize':15})
plt.xlabel("Train types", fontdict={'fontsize':10})
plt.ylabel("Time duration (hrs)", fontdict={'fontsize':10})

plt.style.use("default")
memu = df.loc[df.type == "MEMU"]["total time"]
passe = df.loc[df.type == "Pass"]["total time"]
SF = df.loc[df.type == "SF"]["total time"]
demu = df.loc[df.type == "DEMU"]["total time"]
exp = df.loc[df.type == "Exp"]["total time"]
mail = df.loc[df.type == "Mail"]["total time"]
jshtb =df.loc[df.type == "JShtb"]["total time"] 
Drnt = df.loc[df.type == "Drnt"]["total time"]
Raj = df.loc[df.type == "Raj"]["total time"]
labels = ["memu",'pass',"SF","exp","mail","DEMU","jhstb","drnt","Rajdhani"]
plt.boxplot([memu,passe,SF,exp,mail,demu,jshtb,Drnt,Raj],labels = labels)
plt.savefig("total time_plot.png",dpi=300)
plt.show()
```

```python
plt.title("Distribution of distance travelled by trains in India", fontdict={'fontsize':15})
plt.xlabel("Train types", fontdict={'fontsize':10})
plt.ylabel("Distance (km)", fontdict={'fontsize':10})

plt.style.use("default")
memu = df.loc[df.type == "MEMU"]["distance"]
passe = df.loc[df.type == "Pass"]["distance"]
SF = df.loc[df.type == "SF"]["distance"]
demu = df.loc[df.type == "DEMU"]["distance"]
exp = df.loc[df.type == "Exp"]["distance"]
mail = df.loc[df.type == "Mail"]["distance"]
jshtb =df.loc[df.type == "JShtb"]["distance"] 
Drnt = df.loc[df.type == "Drnt"]["distance"]
Raj = df.loc[df.type == "Raj"]["distance"]
labels = ["memu",'pass',"SF","exp","mail","DEMU","jhstb","drnt","Rajdhani"]
plt.boxplot([memu,passe,SF,exp,mail,demu,jshtb,Drnt,Raj],labels = labels)
plt.savefig("total time_plot.png",dpi=300)
plt.show()
```
