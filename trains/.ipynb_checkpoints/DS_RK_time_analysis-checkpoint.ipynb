```python
#importing libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
df = pd.read_excel('until-totaltime.xlsx')
```

```python
df
```

```python
Arrival_frame=df[['name','arrival']]
Departure_frame=df[['name','departure']]
```

```python
Arrival_frame=Arrival_frame.sort_values(by=['arrival'])
```

```python
Departure_frame=Departure_frame.sort_values(by=['departure'])
```

```python
Arrival_frame
```

```python
Departure_frame
```

```python
Arrival_frame_1=Arrival_frame[(Arrival_frame["arrival"]>='00:00:00') & (Arrival_frame["arrival"]<='04:00:00')]
Arrival_frame_2=Arrival_frame[(Arrival_frame["arrival"]>'04:00:00') & (Arrival_frame["arrival"]<='08:00:00')]
Arrival_frame_3=Arrival_frame[(Arrival_frame["arrival"]>'08:00:00') & (Arrival_frame["arrival"]<='12:00:00')]
Arrival_frame_4=Arrival_frame[(Arrival_frame["arrival"]>'12:00:00') & (Arrival_frame["arrival"]<='16:00:00')]
Arrival_frame_5=Arrival_frame[(Arrival_frame["arrival"]>'16:00:00') & (Arrival_frame["arrival"]<='20:00:00')]
Arrival_frame_6=Arrival_frame[(Arrival_frame["arrival"]>'20:00:00') & (Arrival_frame["arrival"]<'24:00:00')]
```

```python
Departure_frame_1=Departure_frame[(Departure_frame["departure"]>='00:00:00') & (Departure_frame["departure"]<='04:00:00')]
Departure_frame_2=Departure_frame[(Departure_frame["departure"]>'04:00:00') & (Departure_frame["departure"]<='08:00:00')]
Departure_frame_3=Departure_frame[(Departure_frame["departure"]>'08:00:00') & (Departure_frame["departure"]<='12:00:00')]
Departure_frame_4=Departure_frame[(Departure_frame["departure"]>'12:00:00') & (Departure_frame["departure"]<='16:00:00')]
Departure_frame_5=Departure_frame[(Departure_frame["departure"]>'16:00:00') & (Departure_frame["departure"]<='20:00:00')]
Departure_frame_6=Departure_frame[(Departure_frame["departure"]>'20:00:00') & (Departure_frame["departure"]<'24:00:00')]
```

```python
count1=sum([len(Arrival_frame_1),len(Departure_frame_1)])
count2=sum([len(Arrival_frame_2),len(Departure_frame_2)])
count3=sum([len(Arrival_frame_3),len(Departure_frame_3)])
count4=sum([len(Arrival_frame_4),len(Departure_frame_4)])
count5=sum([len(Arrival_frame_5),len(Departure_frame_5)])
count6=sum([len(Arrival_frame_6),len(Departure_frame_6)])
```

```python
count1
```

```python
plt.figure(figsize=(20,10))

values = [count1,count2,count3,count4,count5,count6]
labels = ["Phase-1","Phase-2","Phase-3","Phase-4","Phase-5","Phase-6"]

plt.title("Time wise - train traffic", fontdict={'fontsize':25})
plt.xlabel("Time phase", fontdict={'fontsize':20})
plt.ylabel("Number of trains ", fontdict={'fontsize':20})

plt.plot(labels, values, marker='*', markersize=10, linestyle='-',color='red')

plt.plot([], [], ' ', label="Phase-1 = 00:00:00 hrs to 04:00:00 hrs ")
plt.plot([], [], ' ', label="Phase-2 = 04:00:01 hrs to 08:00:00 hrs")
plt.plot([], [], ' ', label="Phase-3 = 08:00:01 hrs to 12:00:00 hrs ")
plt.plot([], [], ' ', label="Phase-4 = 12:00:01 hrs to 16:00:00 hrs ")
plt.plot([], [], ' ', label="Phase-5 = 16:00:01 hrs to 20:00:00 hrs ")
plt.plot([], [], ' ', label="Phase-6 = 20:00:01 hrs to 23:59:59 hrs ")
plt.legend(fontsize=20)
plt.show()
```
