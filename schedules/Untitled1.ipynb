```python
import pandas as pd
```

```python
df = pd.read_json("schedules.json")
```

```python
df
```

```python
indices = df[(df.arrival == "None") & (df.departure == "None")].index
```

```python
df = df.drop(indices)
```

```python
df
```

```python
df.arrival = df.arrival.replace("None", 0)
```

```python
df.departure = df.departure.replace("None", "Ending")
```

```python
df.head(55)
```

```python
df = df.sort_values(["train_number", "day", "arrival"])
```

```python
df.head(55)
```

```python
import matplotlib.pyplot as plt


plt.figure(figsize=(20,13))
values = [136, 122, 80, 70, 68, 56, 55, 54, 53, 52]
labels = ["HWH", "MSB", "NDLS", "DLI", "MAS", "SBC", "ADI", "LPI", "HYB", "NZM"]

plt.title("Top 10 stations from where trains originate", fontdict={'fontsize': 25})
plt.xlabel("Stations", fontdict={'fontsize': 20})
plt.ylabel("Number of Trains", fontdict={'fontsize': 20})
plt.bar(labels, values)

plt.show()
```

```python
x = df["day"].mean()
y = df["day"].median()

if x > y:
    print("Right skewed")
else:
    print("Left skewed")
print(x,y)
```

```python
plt.figure(figsize=(20,13))

bins = [1, 2, 3, 4]

plt.title("Distribution of Number of days of journey of trains", fontdict={'fontsize':25})
plt.xlabel("Days", fontdict={'fontsize':20})
plt.ylabel("Number of Trains", fontdict={'fontsize':20})

plt.hist(df.day, bins=bins)

plt.show()
```
